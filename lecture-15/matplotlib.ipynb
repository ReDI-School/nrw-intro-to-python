# Matplotlib Intro

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plots
A basic line plot; axis description is generated automatically

```python
fig1, ax = plt.subplots()            # Create a new figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes.

fig2 = plt.figure()                  #alternativly, if you're just working on one axis
plt.plot([1, 2, 3, 4], [1, 4, 2, 3]) #plt.plot always plots to the most recent axis
```

When working outside of the notebook environment (i.e., using .py-files), you need to add `plt.show()`. This opens every figure in a seperate window. The execution of the script will be paused until all opened figures are closed again.

```python
plt.figure() 
plt.title('ordered x-axis')
plt.plot([4, 3, 2, 1], [1, 4, 2, 3])                  # x-values are sorted automatically for numeric axis.

plt.figure()   
plt.title('nominal x-axis')              
plt.plot(['Jan', 'Feb', 'Mar', 'Apr'], [1, 4, 2, 3])  # nominal values (without natural ordering) are possible

plt.figure()    
plt.title('unordered x-axis')               
plt.plot([3, 2, '4', 1], [1, 4, 2, 3])                # no sorting
```

For very simple plots, the declaration of the figure and the axes can be omitted. 

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

A figure can contain multiple plots.

```python
fig1, ax1 = plt.subplots()
plt.plot([1, 2, 3, 4], [1, 4, 2, 3], label="f(x)")  #the label defines the value displayed if the legend is visible
plt.plot([1, 2, 3, 4], [4, 3, 2, 1], label="g(x)")
plt.legend()                                        #display the legend
```

Using the axis object, you can decide to which axis you want to plot. Using the plt object, you always plot to the most recent axis.

```python
fig1, ax1 = plt.subplots()
plt.plot([1, 2, 3, 4], [1, 4, 2, 3], label='a')
plt.plot([1, 2, 3, 4], [4, 3, 2, 1], label='b')
fig2, ax2 = plt.subplots()
plt.plot([1, 2, 3, 4], [1, 3, 3, 4], label='c')
ax1.plot([1, 2, 3, 4], [2, 2, 2, 2], label='d')
ax1.legend()
ax2.legend()
```

### Figure customization
There are a lot of things you might want to change about a plot. Here are the most important methods.

[More about line styles](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
plt.style.use('seaborn')            #add a new styling theme for all plot
                                    #To list all available styles, use: print(plt.style.available)

fig, axis = plt.subplots()          #declare plot
x = np.arange(-3, 3, 0.1)           #create x-values from -3 to 3

#define and plot function number 1
y1 = x**3-4*x*+1            
label1 = '$f(x) = x^3 - 4*x + 1$'
line_style1 = 'ro-'                 #red line with dots
axis.plot(x, y1, line_style1, label=label1)

#define and plot function number 2
y2 = 3*x**2 - 12
label2 = '$g(x) = 3*x^2 - 12$'
line_style2 = 'gv--'                #green dashes with triangles
axis.plot(x, y2, line_style2, label=label2)
#axis.plot(x, y2, label=label2, color='green', marker='v', linestyle='dashed')

axis.legend()                       #show legend
axis.set_ylabel('eggs')             #set labels
axis.set_xlabel('chickens')
axis.set_title('chickens to eggs')
axis.set_xlim(-4,4)                 #manually edit the limits of x und y axes
axis.set_ylim(-12,10)
axis.text(-2, -11, 'no eggs here')  #add some arbitrary text at coordinates
axis.grid(True)                     #add a grid
```

The same can be achieved using the `plt`-object only (that is, without declaring the subplot). Some methods will have other names then.

```python
plt.figure()                        #new figure
x = np.arange(-3, 3, 0.1)          #create x-values from -3 to 3

#define and plot function number 1
y1 = x**3-4*x*+1
label1 = '$f(x) = x^3 - 4*x + 1$'
line_style1 = 'b-.'                #blue dots and dashes
plt.plot(x, y1, line_style1, label=label1)

#define and plot function number 2
y2 = 3*x**2 - 12
label2 = '$g(x) = 3*x^2 - 12$'
line_style2 = 'y*'                  #yellow stars
plt.plot(x, y2, line_style2, label=label2)

plt.legend()                        #show legend
plt.ylabel('eggs')                  #set labels
plt.xlabel('chickens')
plt.title('chickens to eggs')
plt.xlim(-4,4)                      #manually edit the limits of x und y axes
plt.ylim(-12,10)
plt.text(-2, -11, 'no eggs here')   #add some arbitrary text at coordinates
plt.grid(True)                      #add a grid
```

### Excercise

Plot the following functions in the range from 0 to 3 in a single plot:
* $f(x) = 4x^2-5x$
* $g(x) = 3^x$
* $h(x) = 10/(x + 0.5)$

Use different colors for every function and add a legend.

```python
plt.figure()                        #new figure
x = np.arange(0, 3, 0.1)            #create x-values from 0 to 3

#define and plot function number 1
y1 = 4*x**2 - 5*x
label1 = '$f(x) = 4*x^2 - 5x$'
line_style1 = 'b-.'                #blue dots and dashes
plt.plot(x, y1, line_style1, label=label1)

#define and plot function number 2
y2 = 3**x
label2 = '$g(x) = 3^x$'
line_style2 = 'y*'                  #yellow stars
plt.plot(x, y2, line_style2, label=label2)

#define and plot function number 3
y2 = 10/(x + 0.5)
label2 = '$h(x) = 10/(x + 0.5)$'
line_style2 = 'r-v'                  #yellow stars
plt.plot(x, y2, line_style2, label=label2)

plt.legend() 
```

### Subplots
Subplots enable you to arrange your plots in a single figure. There are multiple ways to define your subplots. The first method is to select a subplot with the `plt.subplot` method, which takes three numbers. The first number of the code tells the total number of plots in the vertical dimension, the second is for the horizontal dimension. The third number is the index of the currently selected plot. This is best shown by example. Read more [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html)

```python
plt.figure()

plt.subplot(3, 2, 1)
plt.plot([1,2,3,4], [6,5,4,3])
plt.title('plot 1')
#... do other stuff

plt.subplot(322)    #alternative notation without commas
plt.plot([1,2,3,4], [4,5,6,7])
plt.title('plot 2')

plt.subplot(3, 2, 3)
plt.title('plot 3')

plt.subplot(3, 2, 4)
plt.title('plot 4')

plt.subplot(3, 2, 5)
plt.title('plot 5')


plt.subplot(3, 2, 6)
plt.title('plot 6')

plt.subplots_adjust(hspace=0.8) #adjust the subplots
```

The second option is to create a two-dimensional array of axes. Read more [here](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html).

```python
num_horizontal = 3
num_vertical = 2
fig, axes = plt.subplots(num_vertical, num_horizontal) #axes is a two-dimensional array of the individual axis
print(axes)

x = np.arange(-10, 10, 0.01)
for n, axes_horizontal in enumerate(axes):                          #loop through both dimensions of the axes
  for m, axis in enumerate(axes_horizontal):
    plot_number = n*num_horizontal + m                              
    axis.plot(x, x**plot_number, label=f'$f(x) = x^{plot_number}$') #the enclosing $-signs enable mathematical notation
    axis.set_title(f'plot {plot_number+1}')
    axis.legend()

plt.subplots_adjust(hspace=0.8, wspace=0.8, top=1, right=2)          #adjust the subplots
```

### Scaled y-Axis
In some cases it can be necessary to use other scales than the default linear scale, for example when dealing with exponential data.

```python
# data for the x axis => array from 0 to 10 with steps of 0.01
x = np.arange(0, 10, 0.01)

# plot with various axes scales
plt.figure()

# linear scale, linear function
plt.subplot(221)
plt.plot(x, x, label='f(x) = x')
plt.yscale('linear')
plt.title('linear scale, linear function')
plt.legend()
plt.grid(True)

# linear scale, exponential function
plt.subplot(222)
plt.plot(x, 2**x, label='f(x) = 2^x')
plt.yscale('linear')
plt.title('linear scale, exponential function')
plt.legend()
plt.grid(True)

# logarithmic scale, linear function
plt.subplot(223)
plt.plot(x, x, label='f(x) = x')
plt.yscale('log')
plt.title('logarithmic scale, linear function')
plt.legend()
plt.grid(True)

# logarithmic scale, exponential function
plt.subplot(224)
plt.plot(x, 2**x, label='f(x) = 2^x')
plt.yscale('log')
plt.title('logarithmic scale, exponential function')
plt.legend()
plt.grid(True)

# Adjust the subplot layout
plt.subplots_adjust(top=1.5, bottom=0, left=0.10, right=1.95, hspace=0.25, wspace=0.35)
```

This was really repetitive. Here's the same code using a function:

```python
def plot_with_scale(subplot, x, y, label, yscale, title):
  plt.subplot(subplot)
  plt.plot(x, y, label=label)
  plt.yscale(yscale)
  plt.title(title)
  plt.legend()
  plt.grid(True)

# data for the x axis => array from 0 to 10 with steps of 0.01
x = np.arange(0, 10, 0.01)
plt.figure()

plot_with_scale(221, x, x, '$f(x) = x$', 'linear', 'linear scale, linear function')
plot_with_scale(222, x, 2**x, '$f(x) = 2^x$', 'linear', 'linear scale, exponential function')
plot_with_scale(223, x, x, '$f(x) = x$', 'log', 'logarithmic scale, linear function')
plot_with_scale(224, x, 2**x, '$f(x) = 2^x$', 'log', 'logarithmic scale, exponential function')

# Adjust the subplot layout
plt.subplots_adjust(top=1.5, bottom=0, left=0.10, right=1.95, hspace=0.25, wspace=0.35)
```

### Scatter Plot customization

```python
#generate some random data
#data points are drawn randomly from normal distributions in order to simulate two clusters
number_of_points = 1000
x1 = np.random.normal(3, 1.5, number_of_points)   #cluster 1 x-values
x2 = np.random.normal(0, 0.2, number_of_points)   #cluster 2 x-values
x = np.append(x1, x2)                             
y1 = np.random.normal(4, 1.8, number_of_points)   #cluster 1 y-values
y2 = np.random.normal(0.5, 0.4, number_of_points) #cluster 2 x-values
y = np.append(y1, y2)
sizes = np.random.rand(2*number_of_points)**2*100  #generate random sizes for all data points
color = ['red']*number_of_points + ['green']*number_of_points  #ordered list that assigns a color to every data point

#basic scatter plot with sizes
plt.figure()
plt.title('Scatter Plot with Overlapping Data')
plt.scatter(x, y, s=sizes)
```

```python

#scatter plot with colors and alpha
plt.figure()
plt.title('Transparancy helps to estimate the density of data points')
plt.scatter(x, y, s=sizes, alpha=0.2, c=color)
```

```python
#Using a rule to estimate the clusters - this rule could be the output of an machine learning (AI) model
position_color = ['green' if xi < 0.5 and yi < -0.5*xi + 2 else 'red' for (xi,yi) in zip(x,y)]
plt.subplot(121)
plt.title('Rule-based cluster identification')
plt.scatter(x, y, s=sizes, alpha=0.2, c=position_color)

#Split the data into multiple datasets to add further plot customization
x1 = [xi for (xi,yi) in zip(x,y) if xi < 0.5 and yi < -0.5*xi + 2]
y1 = [yi for (xi,yi) in zip(x,y) if xi < 0.5 and yi < -0.5*xi + 2]
x2 = [xi for (xi,yi) in zip(x,y) if not (xi < 0.5 and yi < -0.5*xi + 2)]
y2 = [yi for (xi,yi) in zip(x,y) if not (xi < 0.5 and yi < -0.5*xi + 2)]
plt.subplot(122)
plt.title('Rule-based cluster identification with different shapes')
plt.scatter(x1, y1, alpha=0.2, c='green', marker='*', label="cluster 1")
plt.scatter(x2, y2, alpha=0.2, c='red', marker='v', label="cluster 2")
plt.legend()

# Adjust the subplot layout
plt.subplots_adjust(top=0.75, bottom=0, left=0.10, right=1.95, hspace=0.25, wspace=0.35)
```

## Mixed Plots

```python
dates = ['Jan', 'Feb', 'Mar', 'Apr', 'Jun', 'Jul']
data = [np.random.rand(5) for month in dates]
mean = [d.mean() for d in data]
maxi = [d.max() for d in data]
stdev = [d.std() for d in data]
plt.figure()
plt.bar(dates, maxi, label="maximum value", width=0.4)
plt.plot(dates, mean, color="red", label="mean value")
plt.legend()
plt.title('A plot using bars and lines at the same time')
```
